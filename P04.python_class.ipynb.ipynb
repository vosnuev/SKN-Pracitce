### 문제 1. 주문 항목 객체 만들기

카페 주문 시스템에서 “주문 항목”을 객체로 표현하려고 합니다.

아래 요구사항을 만족하는 `OrderItem` 클래스를 작성하세요.

1. 속성: `name`, `unit_price`, `qty`
2. 메소드:
    - `total_price()` : `unit_price * qty` 반환
    - `__str__()` : `"아메리카노 x2 = 6000원"` 형태로 반환
3. 실행 예시를 참고하여 출력이 나오도록 테스트 코드도 작성

```
# 기대 출력 예시
# 아메리카노 x2 = 6000원
# 치즈케이크 x1 = 5500원
# 총합: 11500원
```

````python
class OrderItem:
  
  def __init__ (self, name, unit_price, qty):
    self.name = name
    self.unit_price = unit_price
    self.qty = qty
    self.total_price = (unit_price * qty)

  def __str__(self):
    return f"{self.name} * {self.qty} = {self.total_price}원"
  

americano = OrderItem("아메리카노", 3000, 2)
print(americano)
cheese_cake = OrderItem("치즈케이크", 5500, 1)
print(cheese_cake)
total = americano.total_price + cheese_cake.total_price
print(f"총합: {total}원")
````

### 문제 2. 캡슐화 + 검증 로직

`OrderItem`의 `unit_price`와 `qty`는 음수가 될 수 없습니다.

다음 요구사항을 반영해 확장하세요.

1. `unit_price`, `qty`는 외부에서 직접 수정되지 않게(캡슐화) 처리
2. 값을 바꾸려면 `set_unit_price()`, `set_qty()` 메소드를 통해서만 바꿀 수 있게 함
3. 음수 값이 들어오면 값을 변경하지 않고 `"가격/수량은 0 이상이어야 합니다."`를 출력

````python
class OrderItem:
  
  def __init__ (self, name, unit_price, qty):
    self.name = name
    self.__unit_price = unit_price                  # 캡슐화
    self.__qty = qty                                # 캡슐화
    self.total_price = (unit_price * qty)
    
  def set_unit_price(self, price):                  # 메소드 설정
      if price < 0:
          print("가격/수량은 0 이상이어야 합니다.")
      else:
         self.__unit_price = price
         self.total_price = self.__unit_price * self.__qty

  def set_qty(self, qty):                           # 메소드 설정
      if qty < 0:
          print("가격/수량은 0 이상이어야 합니다.")
      else:
         self.__qty = qty
         self.total_price = self.__unit_price * self.__qty

  def __str__(self):
     return f"{self.name} * {self.__qty} = {self.total_price}원"
  

americano = OrderItem("아메리카노", 3000, 2)
print(americano)

americano.set_unit_price(-500)
americano.set_qty(-3)
print(americano)
````

### 문제 3. 상속으로 메뉴 확장

`MenuItem`을 부모 클래스로 두고 `Coffee`, `Dessert`가 상속받도록 구성하세요.

1. `MenuItem` 공통 속성: `name`, `unit_price`
2. `Coffee`는 추가 속성: `shot` (기본 1)
3. `Dessert`는 추가 속성: `size` ("S", "M", "L" 중 하나)
4. `get_info()` 메서드:
    - Coffee: `"아메리카노(샷 2) - 3000원"`
    - Dessert: `"치즈케이크(L) - 5500원"`
5. 다형성: `items = [Coffee(...), Dessert(...)]` 같은 리스트에서 `for`로 `get_info()`를 호출해도 동일하게 동작

###

````python
class MenuItem:
  
  def __init__(self, name, unit_price):
    self.name = name
    self.unit_price = unit_price    

  def get_info(self):
    return f"{self.name} - {self.unit_price}원"
  
# 상속
class Coffee(MenuItem):
  def __init__(self, name, unit_price, shot=1):
    super().__init__(name, unit_price)
    self.shot = shot
  
  def get_info(self):
    return f"{self.name}(샷 {self.shot}) - {self.unit_price}원"
  
class Dessert(MenuItem):
  def __init__(self, name, unit_price, size = "S"):
    super().__init__(name, unit_price)
    self.size = size
  
  def get_info(self):
    return f"{self.name}({self.size}) - {self.unit_price}원"

# 객체
# m1 = Coffee("아메리카노", 3000, 2)
# print(m1)
# m2 = Dessert("치즈케이크", 5500, "L")
# print(m2)

# 반복문
items = [Coffee("아메리카노", 3000, 2),Dessert("치즈케이크", 5500, "L")]
for item in items:
  print(item.get_info())
````

### 문제 4. 장바구니(Cart)로 집계하기

`Cart` 클래스를 만들고 아래 기능을 구현하세요.

1. `add(item: OrderItem)` : 장바구니에 담기
2. `total()` : 장바구니 총합 반환
3. `receipt()` : 영수증 출력 (각 항목 + 총합)

````python
class MenuItem:
  
  def __init__(self, name, unit_price):
    self.name = name
    self.unit_price = unit_price    

  def get_info(self):
    return f"{self.name} - {self.unit_price}원"
  
# 상속
class Coffee(MenuItem):
  def __init__(self, name, unit_price, shot=1):
    super().__init__(name, unit_price)
    self.shot = shot
  
  def get_info(self):
    return f"{self.name}(샷 {self.shot}) - {self.unit_price}원"
  
class Dessert(MenuItem):
  def __init__(self, name, unit_price, size = "S"):
    super().__init__(name, unit_price)
    self.size = size
  
  def get_info(self):
    return f"{self.name}({self.size}) - {self.unit_price}원"

class Cart:
  def __init__(self):
    self.items = []
  
  def add(self, item):
    self.items.append(item)
    print(f"{item.name}이 장바구니에 담겼습니다.")

  def total(self):
    total_sum = 0
    for item in self.items:
        total_sum += item.unit_price
    return total_sum

  def receipt(self):
    print("---영수증---")
    for item in self.items:
      print(item.get_info())

    print("-" * 20)
    print(f"총 합계: {self.total()}원")


i1 = Coffee("아메리카노", 3000, 2)
i2 = Coffee("카페라떼", 4500, 1)
i3 = Dessert("치즈케이크", 5000, "L")

my_cart = Cart()
my_cart.add(i1)
my_cart.add(i2)
my_cart.add(i3)

my_cart.receipt()
````
