## 5. 값 수정
1) 결제상태가 취소인 행의 수량을 0으로 수정하시오.
2) 지역이 결측치인 행의 지역을 '미상'으로 수정하시오.

```python
df = pd.DataFrame({
    '주문일자': ['2026-01-03', '2026-01-05', '2026-01-05', '2026-01-08', '2026-01-10', '2026-01-12', '2026-01-12', '2026-01-15'],
    '고객명': ['김철수', '이영희', '박민수', '최지영', '정다은', '홍길동', '홍길동', '강민호'],
    '지역': ['서울', '부산', '서울', '대전', None, '부산', '부산', '서울'],
    '카테고리': ['전자', '식품', '전자', '패션', '식품', '전자', '전자', '패션'],
    '상품명': ['노트북', '사과', '마우스', '자켓', '우유', '키보드', '키보드', '운동화'],
    '수량': [1, 10, 2, 1, np.nan, 1, 1, 2],
    '단가': [1500000, 3000, 25000, 120000, 2500, 50000, 50000, 90000],
    '결제상태': ['완료', '완료', '완료', '취소', '완료', '완료', '완료', '완료']
},index=['O001', 'O002', 'O003', 'O004', 'O005', 'O006', 'O006', 'O008'],)

df.index.name = ('주문번호')
print(df)
print()
# 1) 결제상태가 취소인 행의 수량을 0으로 수정
df.loc[df['결제상태'] == '취소', '수량'] = 0
df

# 2) 지역이 결측치인 행의 지역을 '미상'으로 수정
df['지역'] = df['지역'].fillna('미상')
df
```
