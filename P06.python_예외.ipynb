### 문제 1. 안전한 정수 입력 함수

사용자에게 숫자를 입력받는 함수 `input_int(prompt)`를 작성하세요.

1. 정수가 아니면 `"정수를 입력하세요."`를 출력하세요.
2. 올바른 정수를 입력할 때까지 반복하세요.
3. 정상 입력 시 `int` 값을 반환하세요.

```python
def input_int(prompt):
    while True:
        try:
            user_input = input(prompt)
            n = int(user_input)
            return n
        except ValueError:
            print("정수만 입력하세요.")
            
if __name__ == "__main__":
    result = input_int("숫자를 입력하세요: ")
    print(f"입력하신 숫자는 {result}입니다.")
```

### 문제 2. 나눗셈 계산기

두 정수를 입력받아 `a / b` 결과를 출력하세요.

1. 0으로 나누면 `"0으로 나눌 수 없습니다."`를 출력하세요.
2. 정수가 아닌 입력이면 `"정수만 입력하세요."`를 출력하세요.
3. 어떤 경우든 프로그램이 종료되지 않고 다시 입력받도록 작성하세요.

```python
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "0으로 나눌 수 없습니다."

def input_int(prompt):
    while True:
        try:
            user_input = input(prompt)
            return int(user_input)
        
        except ValueError:
            print("정수만 입력하세요.")
       
if __name__ == "__main__":
        a = input_int("첫 번째 숫자 (a)를 입력하세요: ")

        while True:
             b = input_int("두 번째 숫자 (b)를 입력하세요: ")
             if b != 0:
                 break
             print("0으로 나눌 수 없습니다. 다시 입력해 주세요")

        result = divide(a, b)  
        print(f"계산 결과는 {result}")
```

### 문제 3. 파일 읽기 예외 처리

`students.txt` 파일을 읽어 학생 점수의 평균을 출력하세요.

조건

1. 파일이 없으면 `"학생 데이터 파일이 없습니다. 먼저 생성하세요."`를 출력하세요.
2. 파일이 있으면 평균을 출력하세요.
3. 예외 처리는 `try / except / else` 구조로 작성하세요.

```python
from pathlib import Path
BASE_DIR = Path(".").resolve()
DATA_DIR = BASE_DIR / "data"
FILE_PATH = DATA_DIR / "students.txt"

class StudentScore:
    def __init__(self, name, subject, score,):
        self.name = name
        self.subject = subject
        self.score = score

def ensure_data_dir():
    DATA_DIR.mkdir(exist_ok=True)

def student_average(file_name):
    try:
        with open(file_name, "r", encoding="utf-8") as f:
            lines = f.readlines()

    except FileNotFoundError:
        print("학생 데이터 파일이 없습니다. 먼저 생성하세요")
        
    else:
        if not lines:
            print("파일은 있지만 ㄷ")
            return
        
        total_score = 0
        count = 0

        for line in lines:
            parts = line.strip().split(",")
            if len(parts) == 2:
                total_score += int(parts[1])
                count += 1

        if count > 0:
            avg= total_score / count
            print(f"평균 점수: {avg}점 (총 {count}명)")

        
if __name__ == "__main__":
    ensure_data_dir ()

    hong = StudentScore("홍길동", "수학", 70)
    sinsa = StudentScore("신사임당", "수학", 90)
    student_list = [hong, sinsa]

    with open(FILE_PATH, "w", encoding="utf-8") as f:
        for s in student_list:
            f.write(f"{s.name},{s.score}\n")

student_average(FILE_PATH)
```

### 문제 4. 사용자 정의 예외로 규칙 강제

점수 입력 규칙: **0~100 사이 정수만 허용**

1. 규칙을 위반하면 `InvalidScoreError` 예외를 발생시키세요.
2. main 로직에서 해당 예외를 잡아 `"점수는 0~100 사이여야 합니다."`를 출력하세요.
3. 올바른 점수를 입력할 때까지 반복하세요.

```python
class InvalidScoreError(Exception):
    pass

def get_valid_number(prompt):
    
    while True:
        try:
            user_input = input(prompt)
            score = int(user_input)
            
            if not (0 <= score <= 100):
                raise InvalidScoreError
            
            return score 
            
        except InvalidScoreError:
            print("[except] 점수는 0~100 사이여야 합니다.")
    
            
if __name__ == "__main__":
    result = get_valid_number("점수를 입력하세요: ")
    print(f"점수는 {result}점 입니다.")
```

### 문제 5. finally 사용하기

프로그램 실행 중 예외가 발생할 수 있는 코드를 작성하세요.

조건

1. 예외가 발생하면 `"오류가 발생했습니다."`를 출력하세요.
2. 어떤 경우든 `"프로그램을 종료합니다."`가 마지막에 출력되도록 `finally`를 사용하세요.

```python
class InvalidScoreError(Exception):
    pass

def get_valid_number(prompt):
    
    while True:
        try:
            user_input = input(prompt)
            score = int(user_input)
            
            if not (0 <= score <= 100):
                raise InvalidScoreError
            
            return score 
            
        except InvalidScoreError:
            print("[except] 오류가 발생했습니다. 0~100 사이의 점수만 가능합니다.")

        finally:
            if (0 <= score <= 100):
                score.close()
            print("[finally] 입력을 확인했습니다. 프로그램을 종료합니다.")
    
            
if __name__ == "__main__":
    result = get_valid_number("점수를 입력하세요: ")
    print(f"입력하신 숫자는 {result}입니다.")
    print(f"입력을 확인했습니다. 프로그램을 종료합니다.")
```
